# Chemical reaction modelling package: examples.

This file show some examples of chemical reaction simulations.

The first step is to import simulation script

```python
import models
%matplotlib notebook
```

## 1. Single reaction:
Irreversible reaction, rate constant is equal 1.
$$
A\rightarrow B
$$
Starting concentration of A=1, B=0. Simulation time = 10 seconds.

```python
chem1 = models.Chemistry() #Creating new chemical simulation domain
chem1.reaction('A=>B') #Adding chemical reaction with default k=1
chem1.initial_concentrations(A=1, B=0)
chem1.time_stop = 10
chem1.run()
```

## 2. System of reactions:
System of reactions, rate constants are equal 1.

Some reactions are reversible:
$$
A\rightarrow B \\
A+X \leftrightarrow C+D \\
B+Y \leftrightarrow C+E \\
E+D \leftrightarrow F
$$
Starting concentrations are: A=1, X=1, B=1, Y=1, the rest is 0. Simulation time = 30 seconds.

```python
chem2 = models.Chemistry() #Creating new empty chemical simulation domain
chem2.reaction('A=>B')
chem2.reaction('A+X<=>C+D')
chem2.reaction('B+Y<=>C+E')
chem2.reaction('E+D<=>F')
chem2.initial_concentrations(A=1, X=1, B=1, Y=1)
chem2.time_stop = 30
chem2.run()
```

## 3. Defining rate constants:
#### Irreversible reaction:

By default, all rate constants are equal 1.

$$
A \rightarrow B, \quad k=0.1 \\
X \rightarrow Y, \quad k=2
$$

```python
chem3 = models.Chemistry()
chem3.reaction('A=>B', k=0.1)
chem3.reaction('X=>Y', k=2)
chem3.time_stop = 50
chem3.initial_concentrations(A=1, X=1)
chem3.run()
```

#### Reversible reaction:

By default, all rate constants are equal 1.

$$
A \leftrightarrow B, \\ \quad k_{forward}=1.5, \\ \quad k_{backward}=1\\
$$

Forward rate constant $k_{forward}$ is set as `k1`

Backward rate constant $k_{backward}$ is set as `k2`

```python
chem4 = models.Chemistry()
chem4.reaction('A<=>B', k1=1.5, k2=1)
chem4.time_stop = 10
chem4.initial_concentrations(A=1)
chem4.run()
```
