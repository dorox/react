```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as p
```

```python
def ode(t,c):
    x = c[0]
    y = c[1]
    a=1
    b=3
    dxdt = a+x**2*y-b*x-x
    dydt = b*x-x**2*y
    return [dxdt, dydt]
```

```python
sol = solve_ivp(ode, (0,100), (1,1), max_step=0.1)
```

```python
t = sol.t
x = sol.y[0]
y = sol.y[1]
```

```python
p.plot(t,x)
p.plot(t,y)
```
