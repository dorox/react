```python
c = react.models.Chemistry()
c.reaction('A=>B')
c.reaction('B=>C')

V=10
r1= react.models.CSTR(V=V)
r1.inlet(A=react.tools.step())
r1.inlet(B=0)

r2 = react.models.CSTR(V=V)
r2.inlet(A=0)
r2.add(c)

r3 = react.models.CSTR(V=V)
r3.inlet(A=0)

f = react.Flowsheet()
f.connect(r1,r2)
f.connect(r2,r3)
f.run()
```
