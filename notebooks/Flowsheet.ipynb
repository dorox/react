```python
V=5
r1= react.models.CSTR(V=V)
r1.inlet(A=react.tools.gaussian())
_=r1.run(False)

r2 = react.models.CSTR(V=V)
r2.inlet(A=0)
_=r2.run(False)

r3 = react.models.CSTR(V=V)
r3.inlet(A=0)
_=r3.run(False)

f = react.Flowsheet()
f.connect(r1,r2)
f.connect(r2,r3)
f.run()
```

```python
c = react.models.Chemistry()
c.reaction('A=>B')

V=5
r1= react.models.CSTR(V=V)
r1.inlet(A=react.tools.rect())
r1.inlet(B=0)

r2 = react.models.CSTR(V=V)
r2.inlet(A=0)
r2.add(c)

r3 = react.models.CSTR(V=V)
r3.inlet(A=0)

f = react.Flowsheet()
f.connect(r1,r2)
f.connect(r2,r3)
f.run()
```

```python
t=r3.solution['t']
y=r3.solution['B']
y2=r3.solution['A']
np.trapz(y,t)+np.trapz(y2,t)
```
