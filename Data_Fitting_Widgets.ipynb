```python
import ipywidgets as widgets
import numpy as np
from ipywidgets import interact, interact_manual
chem2 = models.Chemistry()
chem2.reaction('A+X=>B+Y', k = 1e-10)
chem2.reaction('B+A=>C+Y', k = 1e-10)
chem2.initial_concentrations(A = 1, X =100)
chem2.import_data('data/Expt1.csv')
```

```python
@interact
def show(k1 = (0,1,1e-7), k2 = (0,1,1e-7)):
    chem2.rate_constants = [k1, k2]
    _=chem2.run()
```
